```python
import matplotlib.pyplot as plt

def round(a):
    return int(a + 0.5)

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = abs(dx) if abs(dx) > abs(dy) else abs(dy)
    Xinc = dx / steps
    Yinc = dy / steps
    x, y = x1, y1
    points = []
    for _ in range(steps):
        points.append((round(x), round(y)))
        x += Xinc
        y += Yinc
    return points

def bresenham(a1, b1, a2, b2):
    da = abs(a2 - a1)
    db = abs(b2 - b1)
    slope = db / da if da != 0 else float('inf')
    a, b = a1, b1
    points = []
    if slope <= 1:
        p = 2 * db - da
        for _ in range(da):
            points.append((a, b))
            a += 1
            if p < 0:
                p = p + 2 * db
            else:
                b += 1
                p = p + 2 * db - 2 * da
    else:
        p = 2 * da - db
        for _ in range(db):
            points.append((a, b))
            b += 1
            if p < 0:
                p = p + 2 * da
            else:
                a += 1
                p = p + 2 * da - 2 * db
    return points

def plot_line(points):
    plt.figure(figsize=(5,5))
    plt.plot(*zip(*points), marker='o')
    plt.show()

if __name__ == "__main__":
    x1 = int(input("Enter x1: "))
    y1 = int(input("Enter y1: "))
    x2 = int(input("Enter x2: "))
    y2 = int(input("Enter y2: "))
    print("DDA:")
    points = dda(x1, y1, x2, y2)
    plot_line(points)
    a1 = int(input("Enter x1: "))
    b1 = int(input("Enter y1: "))
    a2 = int(input("Enter x2: "))
    b2 = int(input("Enter y2: "))
    print("Bresenham:")
    points = bresenham(a1, b1, a2, b2)
    plot_line(points)
```
